### GIẢI THÍCH CÁC CHỈ SỐ CÓ TRONG DATA:
#### 1. Đường Trung Bình Động Lũy Thừa (Exponential Moving Average - EMA)

EMA là một loại đường trung bình động mà đặt nặng hơn vào các điểm dữ liệu gần nhất (ngày gần nhất). Điều này làm cho EMA phản ứng nhanh hơn với các thay đổi giá so với Đường Trung Bình Động Đơn Giản (SMA).

-   **Cách dùng**:
    * Xác định xu hướng: Khi giá nằm trên EMA, đó có thể là xu hướng tăng; ngược lại là xu hướng giảm.
    * Tín hiệu giao cắt: Giao cắt giữa các đường EMA ngắn hạn và dài hạn (ví dụ: EMA 20 cắt lên EMA 50) có thể tạo tín hiệu mua/bán.
    * Mức hỗ trợ/kháng cự động.

-   **Các cột đã tạo**:
    * `EMA_20`: Đường EMA 20 kỳ (thường là 20 ngày).
    * `EMA_50`: Đường EMA 50 kỳ.
    * `EMA_200`: Đường EMA 200 kỳ (thường được coi là đường xu hướng dài hạn quan trọng).

#### 2. Chỉ Số Sức Mạnh Tương Đối (Relative Strength Index - RSI)

RSI là một chỉ báo động lượng đo lường tốc độ và sự thay đổi của các biến động giá. RSI dao động trong khoảng từ 0 đến 100.

-   **Cách dùng**:
    * **Quá mua/Quá bán**:
        * Giá trị RSI trên 70 thường cho thấy tình trạng quá mua (overbought), có khả năng giá sẽ điều chỉnh giảm.
        * Giá trị RSI dưới 30 thường cho thấy tình trạng quá bán (oversold), có khả năng giá sẽ phục hồi tăng.
    * **Phân kỳ**: Khi hướng di chuyển của RSI ngược lại với hướng di chuyển của giá, nó có thể báo hiệu sự đảo chiều xu hướng tiềm năng.

-   **Cột đã tạo**:
    * `RSI_14`: Chỉ số RSI được tính toán dựa trên chu kỳ 14 kỳ (thường là 14 ngày).

#### 3. Dải Bollinger (Bollinger Bands)

Dải Bollinger bao gồm ba đường được vẽ liên quan đến giá của một tài sản: một đường giữa (thường là SMA), một dải trên và một dải dưới (thường cách đường giữa một số độ lệch chuẩn nhất định).

-   **Cách dùng**:
    * Đo lường biến động: Dải Bollinger mở rộng khi biến động tăng và thu hẹp khi biến động giảm.
    * Tín hiệu quá mua/quá bán: Giá chạm vào dải trên có thể cho thấy tình trạng quá mua, chạm dải dưới có thể cho thấy quá bán. Tuy nhiên, đây không phải là tín hiệu mua/bán trực tiếp mà cần kết hợp với các yếu tố khác.
    * "Thắt cổ chai" (Squeeze): Khi dải Bollinger thu hẹp đáng kể, nó thường báo hiệu một giai đoạn biến động mạnh sắp xảy ra.

-   **Các cột đã tạo** (với chu kỳ 20 và 2 độ lệch chuẩn):
    * `BBM_20_2` (Middle Band): Đường giữa, là SMA 20 kỳ của giá.
    * `BBU_20_2` (Upper Band): Dải trên, được tính bằng Đường giữa + (2 * Độ lệch chuẩn 20 kỳ của giá).
    * `BBL_20_2` (Lower Band): Dải dưới, được tính bằng Đường giữa - (2 * Độ lệch chuẩn 20 kỳ của giá).
    * `BBB_20_2` (Bandwidth): Độ rộng Dải Bollinger, tính bằng `(BBU - BBL) / BBM`. Cho thấy mức độ biến động tương đối.
    * `BBP_20_2` (Percent B): Vị trí của giá đóng cửa so với các dải, tính bằng `(Giá đóng cửa - BBL) / (BBU - BBL)`. Giúp xác định điều kiện quá mua/quá bán.

### 4. Tỷ Suất Sinh Lợi Trễ (Lagged Returns)

Đây là tỷ suất sinh lợi của tài sản được tính toán so với một số kỳ trước đó.

-   **Cách dùng**:
    * Đo lường hiệu suất quá khứ trong một khoảng thời gian nhất định.
    * Có thể được sử dụng làm đặc trưng đầu vào cho các mô hình dự đoán.

-   **Các cột đã tạo**:
    * `Return_1`: Tỷ suất sinh lợi của ngày hiện tại so với giá của 1 ngày trước đó. `(Giá[t] / Giá[t-1]) - 1`.
    * `Return_5`: Tỷ suất sinh lợi của ngày hiện tại so với giá của 5 ngày trước đó. `(Giá[t] / Giá[t-5]) - 1`.

### 5. Đặc Trưng Cửa Sổ Trượt (Rolling Window Features)

Các đặc trưng này được tính toán dựa trên một cửa sổ dữ liệu trượt qua thời gian. Ví dụ, với cửa sổ 20 ngày, tại mỗi ngày, chúng ta sẽ nhìn lại 20 ngày gần nhất (bao gồm cả ngày hiện tại) để tính toán.

-   **Cách dùng**:
    * Làm mịn dữ liệu và xác định xu hướng ngắn hạn/trung hạn (ví dụ: rolling mean).
    * Đo lường biến động cục bộ (ví dụ: rolling standard deviation).
    * Xác định các mức giá cao nhất/thấp nhất cục bộ (ví dụ: rolling min/max).

-   **Các cột đã tạo** (`V_` cho cột 'Vol', cùng với kích thước cửa sổ 20):
    * **Từ cột 'Vol' (Khối lượng giao dịch)**:
        *`V_Roll_MEAN_20`: Trung bình trượt của khối lượng giao dịch trong 20 kỳ.

#### 6. Biến Mục Tiêu cho Dự Đoán Phân Loại (Target Variable for Classification)

Để huấn luyện một mô hình học máy dự đoán xem giá sẽ "lên" hay "xuống", chúng ta cần một cột dữ liệu cụ thể cho mô hình biết được kết quả đúng trong quá khứ là gì. Cột này được gọi là **biến mục tiêu**.

-   **Mục đích**: Biến mục tiêu xác định điều mà chúng ta muốn mô hình dự đoán. Trong trường hợp này, đó là hướng di chuyển của giá trong tương lai gần.

-   **Cách tạo**:
    * Chúng ta đã tạo một cột mục tiêu dựa trên việc so sánh giá đóng cửa (`Price`) của ngày hiện tại (`t`) với giá đóng cửa của một ngày trong tương lai (`t+1`).
    * Cụ thể, nếu giá đóng cửa của ngày hôm sau (`Price[t+1]`) cao hơn giá đóng cửa của ngày hiện tại (`Price[t]`), biến mục tiêu sẽ nhận giá trị **1** (biểu thị "lên").
    * Nếu giá đóng cửa của ngày hôm sau thấp hơn hoặc bằng giá đóng cửa của ngày hiện tại, biến mục tiêu sẽ nhận giá trị **0** (biểu thị "xuống" hoặc "không đổi").

-   **Cột đã tạo (ví dụ với `future_periods_for_target = 1`)**:
    * `Target_Up_Next1D`:
        * Giá trị là **1** nếu giá đóng cửa của ngày hôm sau (`Price[t+1]`) cao hơn giá đóng cửa của ngày hôm nay (`Price[t]`).
        * Giá trị là **0** nếu giá đóng cửa của ngày hôm sau (`Price[t+1]`) thấp hơn hoặc bằng giá đóng cửa của ngày hôm nay (`Price[t]`).

## 3.2 Technical indicators

```python
import numpy as np
import pandas_ta as ta
```

```python
ema_periods_to_calc = [20, 50, 200]  # Các chu kỳ EMA mong muốn
rsi_period_to_calc = 14              # Chu kỳ RSI
bollinger_period_to_calc = 20        # Chu kỳ Bollinger Bands
bollinger_std_dev_to_calc = 2        # Độ lệch chuẩn cho Bollinger Bands
```

```python
import pandas as pd
import pandas_ta as ta
import os

# --- Phần 1: Hàm tiện ích để lưu hoặc cập nhật tệp tín hiệu (ĐÃ CẬP NHẬT) ---
def save_or_update_signal_file(asset_name, original_df, new_indicators_df_with_date):
    """
    Lưu hoặc cập nhật tệp CSV tín hiệu, đảm bảo tất cả các cột gốc được bao gồm.
    - asset_name: Tên tài sản.
    - original_df: DataFrame gốc đầy đủ từ tệp _cleaned.csv.
    - new_indicators_df_with_date: DataFrame chứa cột 'Date' và các cột chỉ báo MỚI cần thêm/cập nhật.
    """
    output_filename = f"{asset_name}_signal.csv"

    # 0. Đảm bảo new_indicators_df_with_date có cột 'Date' và xử lý nó
    if 'Date' not in new_indicators_df_with_date.columns:
        print(f"LƯU Ý: Dữ liệu chỉ báo mới cho {asset_name} thiếu cột 'Date'. Không thể lưu.")
        return
    new_indicators_df_with_date['Date'] = pd.to_datetime(new_indicators_df_with_date['Date'])
    new_indicators_indexed = new_indicators_df_with_date.set_index('Date')

    # 1. Bắt đầu với bản sao đầy đủ của dữ liệu gốc, đặt 'Date' làm index
    final_df_indexed = original_df.copy()
    if not isinstance(final_df_indexed.index, pd.DatetimeIndex) or 'Date' in final_df_indexed.columns:
        # Nếu 'Date' là cột, đảm bảo nó là datetime và đặt làm index
        final_df_indexed['Date'] = pd.to_datetime(final_df_indexed['Date'])
        final_df_indexed = final_df_indexed.set_index('Date')
    # Nếu 'Date' đã là index và là DatetimeIndex, không cần làm gì thêm cho bước này

    # 2. Nếu tệp tín hiệu đã tồn tại, đọc và hợp nhất các chỉ báo đã có từ trước
    if os.path.exists(output_filename):
        try:
            existing_signal_df = pd.read_csv(output_filename, parse_dates=['Date'])
            existing_signal_df['Date'] = pd.to_datetime(existing_signal_df['Date'])
            existing_signal_df = existing_signal_df.set_index('Date')
            
            # Thêm các cột từ existing_signal_df mà không có trong original_df (đây là các chỉ báo cũ)
            for col_name in existing_signal_df.columns:
                if col_name not in final_df_indexed.columns: # Chỉ thêm nếu là cột mới (chỉ báo)
                    final_df_indexed[col_name] = existing_signal_df[col_name]
                # Các cột gốc đã có trong final_df_indexed từ original_df
                # Nếu cột chỉ báo đã có trong existing_signal_df và cũng có trong new_indicators_indexed (sắp được thêm),
                # nó sẽ được ghi đè bởi new_indicators_indexed ở bước sau.
        except Exception as e:
            print(f"LỖI khi đọc tệp hiện có {output_filename}: {e}. Sẽ xây dựng lại từ dữ liệu gốc và chỉ báo mới.")
            # final_df_indexed đã chứa dữ liệu gốc, tiếp tục để thêm chỉ báo mới.

    # 3. Thêm hoặc cập nhật các chỉ báo MỚI (từ new_indicators_indexed)
    for col_name in new_indicators_indexed.columns:
        final_df_indexed[col_name] = new_indicators_indexed[col_name]

    # Chuyển 'Date' từ index trở lại thành cột
    final_df = final_df_indexed.reset_index()

    # 4. Sắp xếp lại thứ tự các cột
    # Ưu tiên: Date, sau đó các cột gốc, sau đó các chỉ báo đã tính
    original_cols_expected_order = ['Open', 'High', 'Low', 'Price', 'Vol', 'Change'] # Các cột gốc chính
    
    ordered_cols = []
    if 'Date' in final_df.columns:
        ordered_cols.append('Date')
    
    # Thêm các cột gốc theo thứ tự mong muốn (nếu có)
    for col in original_cols_expected_order:
        if col in final_df.columns and col not in ordered_cols:
            ordered_cols.append(col)
            
    # Thêm các cột gốc khác còn lại (nếu có và chưa được thêm)
    for col in original_df.columns:
        if col not in ordered_cols and col.lower() != 'date':
            ordered_cols.append(col)

    # Thêm các cột chỉ báo (những cột không có trong ordered_cols) và sắp xếp theo tên
    indicator_cols = [col for col in final_df.columns if col not in ordered_cols]
    indicator_cols.sort()
    ordered_cols.extend(indicator_cols)
    
    # Đảm bảo không có cột nào bị trùng lặp trong ordered_cols và tất cả các cột đều có mặt
    final_df = final_df[list(pd.Series(ordered_cols).drop_duplicates())]


    # 5. Loại bỏ các hàng mà TẤT CẢ các cột chỉ báo được TÍNH TOÁN thực sự là NaN
    original_column_names_for_check = [col for col in original_df.columns if col.lower() != 'date']
    calculated_indicator_cols = [col for col in final_df.columns if col.lower() != 'date' and col not in original_column_names_for_check]
    
    if calculated_indicator_cols: # Chỉ thực hiện nếu có cột chỉ báo được tính toán
        final_df.dropna(subset=calculated_indicator_cols, how='all', inplace=True)

    # 6. Lưu tệp
    final_df.to_csv(output_filename, index=False, float_format='%.5f')
    print(f"Đã cập nhật và lưu tệp (bao gồm tất cả cột gốc): {output_filename}")
    print("5 dòng đầu của tệp:")
    print(final_df.head())
    print("-" * 20)

# --- Phần 2: Các hàm tính toán và lưu từng loại chỉ báo ---

# 2.1. Tính toán và lưu các đường EMA
def calculate_and_save_emas(dataframes_dict, ema_periods_list):
    if not dataframes_dict: 
        print("Từ điển dataframes rỗng, không thể tính EMA.")
        return
    print("\n--- Bắt đầu tính toán và lưu các đường EMA ---")
    for asset_name, original_df in dataframes_dict.items():
        if original_df.empty or 'Price' not in original_df.columns or 'Date' not in original_df.columns:
            print(f"Bỏ qua EMA cho {asset_name} do dữ liệu gốc không hợp lệ (thiếu Date/Price hoặc rỗng).")
            continue

        print(f"Tính EMA cho {asset_name}...")
        # DataFrame tạm thời chỉ chứa 'Date' và các chỉ báo EMA mới
        ema_indicators_temp = pd.DataFrame({'Date': original_df['Date']}) 
        
        for period in ema_periods_list:
            ema_indicators_temp[f'EMA_{period}'] = ta.ema(original_df['Price'], length=period)
        
        save_or_update_signal_file(asset_name, original_df, ema_indicators_temp)
    print("--- Hoàn thành việc tính toán và lưu EMAs ---")

# 2.2. Tính toán và lưu RSI
def calculate_and_save_rsi(dataframes_dict, rsi_period_val):
    if not dataframes_dict:
        print("Từ điển dataframes rỗng, không thể tính RSI.")
        return
    print("\n--- Bắt đầu tính toán và lưu RSI ---")
    for asset_name, original_df in dataframes_dict.items():
        if original_df.empty or 'Price' not in original_df.columns or 'Date' not in original_df.columns:
            print(f"Bỏ qua RSI cho {asset_name} do dữ liệu gốc không hợp lệ.")
            continue
            
        print(f"Tính RSI cho {asset_name}...")
        rsi_indicator_temp = pd.DataFrame({'Date': original_df['Date']})
        rsi_indicator_temp[f'RSI_{rsi_period_val}'] = ta.rsi(original_df['Price'], length=rsi_period_val)
        
        save_or_update_signal_file(asset_name, original_df, rsi_indicator_temp)
    print("--- Hoàn thành việc tính toán và lưu RSI ---")

# 2.3. Tính toán và lưu Bollinger Bands
def calculate_and_save_bbands(dataframes_dict, bbands_period_val, bbands_std_dev_val):
    if not dataframes_dict:
        print("Từ điển dataframes rỗng, không thể tính Bollinger Bands.")
        return
    print("\n--- Bắt đầu tính toán và lưu Bollinger Bands ---")
    for asset_name, original_df in dataframes_dict.items():
        if original_df.empty or 'Price' not in original_df.columns or 'Date' not in original_df.columns:
            print(f"Bỏ qua Bollinger Bands cho {asset_name} do dữ liệu gốc không hợp lệ.")
            continue

        print(f"Tính Bollinger Bands cho {asset_name}...")
        bbands_indicators_temp = pd.DataFrame({'Date': original_df['Date']})
        
        # Tính toán Bollinger Bands từ pandas_ta
        temp_bbands_df = ta.bbands(original_df['Price'], length=bbands_period_val, std=bbands_std_dev_val)
        
        if temp_bbands_df is not None and not temp_bbands_df.empty:
            # Gán các cột từ temp_bbands_df vào bbands_indicators_temp
            # Tên cột trong temp_bbands_df thường là: BBL_length_std, BBM_length_std, BBU_length_std, BBB_length_std, BBP_length_std
            bbands_indicators_temp[f'BBL_{bbands_period_val}_{bbands_std_dev_val}'] = temp_bbands_df.iloc[:, 0] # Lower
            bbands_indicators_temp[f'BBM_{bbands_period_val}_{bbands_std_dev_val}'] = temp_bbands_df.iloc[:, 1] # Middle
            bbands_indicators_temp[f'BBU_{bbands_period_val}_{bbands_std_dev_val}'] = temp_bbands_df.iloc[:, 2] # Upper
            if temp_bbands_df.shape[1] > 3: # Bandwidth
                bbands_indicators_temp[f'BBB_{bbands_period_val}_{bbands_std_dev_val}'] = temp_bbands_df.iloc[:, 3]
            if temp_bbands_df.shape[1] > 4: # Percent
                bbands_indicators_temp[f'BBP_{bbands_period_val}_{bbands_std_dev_val}'] = temp_bbands_df.iloc[:, 4]
        else:
            print(f"Không thể tính Bollinger Bands cho {asset_name}. Các cột BB sẽ là NA.")
            # Tạo các cột với giá trị NA nếu không tính được, để đảm bảo cấu trúc file
            bbands_indicators_temp[f'BBL_{bbands_period_val}_{bbands_std_dev_val}'] = pd.NA 
            bbands_indicators_temp[f'BBM_{bbands_period_val}_{bbands_std_dev_val}'] = pd.NA
            bbands_indicators_temp[f'BBU_{bbands_period_val}_{bbands_std_dev_val}'] = pd.NA
            bbands_indicators_temp[f'BBB_{bbands_period_val}_{bbands_std_dev_val}'] = pd.NA
            bbands_indicators_temp[f'BBP_{bbands_period_val}_{bbands_std_dev_val}'] = pd.NA

        save_or_update_signal_file(asset_name, original_df, bbands_indicators_temp)
    print("--- Hoàn thành việc tính toán và lưu Bollinger Bands ---")

#Phần 3: Thực thi

if dataframes:
    # 1. Tính và lưu EMAs
    calculate_and_save_emas(dataframes, ema_periods_to_calc)

    # 2. Tính và lưu RSI (bạn có thể chạy khối này sau, hoặc comment lại nếu chưa muốn)
    calculate_and_save_rsi(dataframes, rsi_period_to_calc)

    # 3. Tính và lưu Bollinger Bands (tương tự, có thể chạy sau hoặc comment lại)
    calculate_and_save_bbands(dataframes, bollinger_period_to_calc, bollinger_std_dev_to_calc)
    
    print("\n--- TẤT CẢ QUÁ TRÌNH HOÀN TẤT ---")
else:
    print("LỖI: Dữ liệu gốc (dataframes) chưa được nạp hoặc bị lỗi. Không thể thực hiện tính toán chỉ báo.")
```

## 3.3  Lagged returns

```python
def calculate_and_save_lagged_returns(dataframes_dict, lags_list):
    """
    Tính toán và lưu các cột tỷ suất sinh lợi trễ.
    - dataframes_dict: Từ điển chứa các DataFrame gốc của tài sản.
    - lags_list: Danh sách các độ trễ cần tính (ví dụ: [1, 5]).
    """
    if not dataframes_dict:
        print("Từ điển dataframes rỗng, không thể tính Lagged Returns.")
        return
    print("\n--- Bắt đầu tính toán và lưu Tỷ suất sinh lợi trễ ---")
    for asset_name, original_df in dataframes_dict.items():
        if original_df.empty or 'Price' not in original_df.columns or 'Date' not in original_df.columns:
            print(f"Bỏ qua Lagged Returns cho {asset_name} do dữ liệu gốc không hợp lệ.")
            continue
            
        print(f"Tính Lagged Returns cho {asset_name}...")
        # DataFrame tạm thời chỉ chứa 'Date' và các cột return mới
        lagged_returns_temp = pd.DataFrame({'Date': original_df['Date']})
        
        for lag in lags_list:
            # Sử dụng pct_change() để tính tỷ suất sinh lợi so với 'lag' kỳ trước đó
            # (price[t] / price[t-lag]) - 1
            lagged_returns_temp[f'Return_{lag}'] = original_df['Price'].pct_change(periods=lag) 
        
        save_or_update_signal_file(asset_name, original_df, lagged_returns_temp)
    print("--- Hoàn thành việc tính toán và lưu Tỷ suất sinh lợi trễ ---")
```

```python
lags_for_returns = [1, 5] 
calculate_and_save_lagged_returns(dataframes, lags_for_returns)
```

## 3.4  Rolling window for Volume

```python
def calculate_and_save_rolling_features(dataframes_dict, window_sizes_list, column_configs):
    """
    Tính toán và lưu các đặc trưng dựa trên cửa sổ trượt.
    - dataframes_dict: Từ điển chứa các DataFrame gốc của tài sản.
    - window_sizes_list: Danh sách các kích thước cửa sổ (ví dụ: [5, 20]).
    - column_configs: Danh sách các tuple, mỗi tuple_mô tả (tên_cột_gốc, list_hàm_rolling, tiền_tố_tên_mới)
                      Ví dụ: [('Price', ['mean'], 'Price'), ('Vol', ['mean'], 'Vol')]
    """
    if not dataframes_dict:
        print("Từ điển dataframes rỗng, không thể tính Rolling Features.")
        return
    print("\n--- Bắt đầu tính toán và lưu các Đặc trưng Cửa sổ Trượt ---")

    for asset_name, original_df in dataframes_dict.items():
        if original_df.empty or 'Date' not in original_df.columns:
            print(f"Bỏ qua Rolling Features cho {asset_name} do dữ liệu gốc không hợp lệ (thiếu Date hoặc rỗng).")
            continue
            
        print(f"Tính Rolling Features cho {asset_name}...")
        rolling_features_temp = pd.DataFrame({'Date': original_df['Date']}) # Bắt đầu với cột Date

        for original_col_name, agg_funcs, new_col_prefix in column_configs:
            if original_col_name not in original_df.columns:
                print(f"LƯU Ý: Cột '{original_col_name}' không tồn tại trong dữ liệu của {asset_name}. Bỏ qua rolling features cho cột này.")
                continue

            for window in window_sizes_list:
                for func_name in agg_funcs:
                    new_feature_col_name = f"{new_col_prefix}_Roll_{func_name.upper()}_{window}"
                    try:
                        if func_name == 'mean':
                            rolling_series = original_df[original_col_name].rolling(window=window, min_periods=1).mean()
                        else:
                            print(f"LƯU Ý: Hàm rolling '{func_name}' chưa được hỗ trợ. Bỏ qua.")
                            continue
                        rolling_features_temp[new_feature_col_name] = rolling_series
                    except Exception as e:
                        print(f"Lỗi khi tính {new_feature_col_name} cho {asset_name}: {e}")
        
        if len(rolling_features_temp.columns) > 1: # Chỉ lưu nếu có ít nhất một feature được tính (ngoài Date)
            save_or_update_signal_file(asset_name, original_df, rolling_features_temp)
        else:
            print(f"Không có rolling features nào được tính cho {asset_name}.")
            
    print("--- Hoàn thành việc tính toán và lưu các Đặc trưng Cửa sổ Trượt ---")
```

```python
rolling_window_sizes = [20]
rolling_column_configurations = [
    ('Vol', ['mean'], 'V')
]
```

```python
if dataframes:
    calculate_and_save_rolling_features(dataframes, rolling_window_sizes, rolling_column_configurations)
    print("\n--- TẤT CẢ QUÁ TRÌNH HOÀN TẤT ---")
```

## 3.5 Target prediction indicator

```python
def save_or_update_signal_file(asset_name, original_df, new_indicators_df_with_date):
    output_filename = f"{asset_name}_signal.csv"
    if 'Date' not in new_indicators_df_with_date.columns:
        print(f"LƯU Ý: Dữ liệu chỉ báo mới cho {asset_name} thiếu cột 'Date'. Không thể lưu.")
        return
    new_indicators_df_with_date['Date'] = pd.to_datetime(new_indicators_df_with_date['Date'])
    new_indicators_indexed = new_indicators_df_with_date.set_index('Date')
    final_df = None
    if os.path.exists(output_filename):
        try:
            current_signal_df = pd.read_csv(output_filename, parse_dates=['Date'])
            current_signal_df['Date'] = pd.to_datetime(current_signal_df['Date'])
            current_signal_df = current_signal_df.set_index('Date')
            for col_name in new_indicators_indexed.columns:
                current_signal_df[col_name] = new_indicators_indexed[col_name]
            final_df = current_signal_df.reset_index()
        except pd.errors.EmptyDataError:
            print(f"LƯU Ý: Tệp {output_filename} tồn tại nhưng rỗng. Sẽ tạo lại từ đầu.")
        except Exception as e:
            print(f"LỖI khi đọc tệp {output_filename}: {e}. Sẽ cố gắng tạo/ghi đè.")
    if final_df is None:
        if not ('Date' in original_df.columns or isinstance(original_df.index, pd.DatetimeIndex)):
            print(f"LƯU Ý: Dữ liệu gốc cho {asset_name} thiếu cột 'Date'. Không thể tạo tệp tín hiệu.")
            return
        base_data = pd.DataFrame()
        if isinstance(original_df.index, pd.DatetimeIndex) and 'Date' not in original_df.columns:
            base_data['Date'] = original_df.index
        else:
            base_data['Date'] = original_df['Date']
        if 'Price' in original_df.columns:
            base_data['Price'] = original_df['Price'].values # Thêm cột Price gốc
        # Thêm các cột gốc khác từ original_df vào base_data
        for col in original_df.columns:
            if col not in base_data.columns and col.lower() != 'date': # Tránh thêm lại Date hoặc Price nếu đã có
                 base_data[col] = original_df[col].values

        base_data['Date'] = pd.to_datetime(base_data['Date'])
        base_data = base_data.set_index('Date')
        final_df = base_data.join(new_indicators_indexed, how='left')
        final_df = final_df.reset_index()

    original_cols_expected_order = ['Open', 'High', 'Low', 'Price', 'Vol', 'Change']
    ordered_cols = []
    if 'Date' in final_df.columns: ordered_cols.append('Date')
    for col in original_cols_expected_order:
        if col in final_df.columns and col not in ordered_cols: ordered_cols.append(col)
    for col in original_df.columns: # Đảm bảo tất cả cột gốc có mặt
        if col not in ordered_cols and col.lower() != 'date': ordered_cols.append(col)
    indicator_cols = [col for col in final_df.columns if col not in ordered_cols]
    indicator_cols.sort()
    ordered_cols.extend(indicator_cols)
    # Xử lý trường hợp cột target có thể nằm trong indicator_cols, di chuyển về cuối nếu muốn
    if 'Target_Up_Next1D' in ordered_cols: # Ví dụ tên cột target
        ordered_cols.remove('Target_Up_Next1D')
        ordered_cols.append('Target_Up_Next1D')
    final_df = final_df[list(pd.Series(ordered_cols).drop_duplicates())]
    original_column_names_for_check = [col for col in original_df.columns if col.lower() != 'date']
    calculated_indicator_cols = [col for col in final_df.columns if col.lower() != 'date' and col not in original_column_names_for_check and not col.startswith('Target_')]
    if calculated_indicator_cols:
        final_df.dropna(subset=calculated_indicator_cols, how='all', inplace=True)
    final_df.to_csv(output_filename, index=False, float_format='%.5f')
    print(f"Đã cập nhật và lưu tệp (bao gồm tất cả cột gốc): {output_filename}")
    print("5 dòng đầu của tệp:")
    print(final_df.head())
    print("-" * 20)

def create_and_save_target_variable(dataframes_dict, future_periods=1):
    """
    Tạo và lưu biến mục tiêu cho dự đoán lên/xuống.
    - dataframes_dict: Từ điển chứa các DataFrame gốc (đã có cột 'Price' và 'Date').
    - future_periods: Số kỳ trong tương lai để so sánh giá (mặc định là 1).
    """
    if not dataframes_dict:
        print("Từ điển dataframes rỗng, không thể tạo biến mục tiêu.")
        return
    print(f"\n--- Bắt đầu tạo biến mục tiêu (cho {future_periods} kỳ tới) ---")

    for asset_name, original_df in dataframes_dict.items():
        if original_df.empty or 'Price' not in original_df.columns or 'Date' not in original_df.columns:
            print(f"Bỏ qua tạo biến mục tiêu cho {asset_name} do dữ liệu gốc không hợp lệ.")
            continue
            
        print(f"Tạo biến mục tiêu cho {asset_name}...")
        
        target_df = pd.DataFrame({'Date': original_df['Date']})
        
        # Tính giá trong tương lai bằng cách dịch chuyển (shift) cột 'Price' ngược lại N kỳ
        future_price = original_df['Price'].shift(-future_periods)
        
        # Tạo mục tiêu: 1 nếu giá tương lai > giá hiện tại, 0 nếu ngược lại (hoặc NaN nếu không có giá tương lai)
        target_df[f'Target_Up_Next{future_periods}D'] = np.where(future_price > original_df['Price'], 1, 
                                                                 np.where(future_price <= original_df['Price'], 0, np.nan))
        # Gọi hàm lưu trữ (hàm này sẽ hợp nhất cột target mới vào file _signal.csv)
        save_or_update_signal_file(asset_name, original_df, target_df)
        
    print("--- Hoàn thành việc tạo và lưu biến mục tiêu ---")


# MỚI: Tham số cho biến mục tiêu
future_periods_for_target = 1


# === THỰC THI ===
if dataframes: # Chỉ chạy nếu dataframes được nạp thành công
    # MỚI: Tạo và lưu biến mục tiêu
    create_and_save_target_variable(dataframes, future_periods_for_target)
    
    print("\n--- QUÁ TRÌNH TẠO BIẾN MỤC TIÊU HOÀN TẤT ---")
else:
    print("LỖI: Dữ liệu gốc (dataframes) chưa được nạp hoặc bị lỗi. Không thể thực hiện.")
```
