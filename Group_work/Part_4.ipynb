## 4.1 Auto Arima

```python
# PART 4: Time Series Forecasting using ARIMA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import traceback
import os

# File paths (use the signal files generated in Part 3)
base_path = "C:/Users/Dell/Documents/(1) Neu/Big data/Code.py"
assets = {
    "BTC": os.path.join(base_path, "BTC_signal.csv"),
    "VNI": os.path.join(base_path, "VNI_signal.csv"),
    "XAU": os.path.join(base_path, "XAU_signal.csv")
}

# Parameters
forecast_horizon = 1        # Predict next day log return
initial_train_size = 60     # Initial training window (e.g., 60 days)


def forecast_asset(name, path, is_ARIMA=True):
    print(f"\n=== Processing {name} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)

    # Compute log returns
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)

    preds = []
    true_vals = []
    dates = []

    try:
        if is_ARIMA:
            # Automatic tuning of the ARIMA model
            train_data = df['Log_Return'].iloc[:initial_train_size]
            model_auto = auto_arima(
                train_data.values,
                start_p=4,
                start_q=4,
                test='adf',
                max_p=5, max_q=5,
                m=1,
                d=None,
                seasonal=False,
                start_P=0,
                D=0,
                start_Q=0,
                trace=True,
                error_action='ignore',
                suppress_warnings=False,
                stepwise=True
            )
            print(model_auto.summary())
            arima_order = model_auto.order
            print(f"[{name}] Optimal ARIMA order: {arima_order}")

            # Plot diagnostics once
            fig = model_auto.plot_diagnostics(figsize=(12, 10))
            plt.suptitle(f"{name} - Auto ARIMA Diagnostics", fontsize=14)
            plt.show()

            # Expanding window forecast using determined order, forecast every 10th point
            for i in range(initial_train_size, len(df) - forecast_horizon, 10):
                train_data = df['Log_Return'].iloc[:i]
                test_actual = df['Log_Return'].iloc[i + forecast_horizon - 1]
                test_date = df.index[i + forecast_horizon - 1]

                model = ARIMA(train_data, order=arima_order)
                model_fit = model.fit()
                forecast = model_fit.forecast(steps=forecast_horizon)[-1]
                preds.append(forecast)
                true_vals.append(test_actual)
                dates.append(test_date)
        else:
            print("ARIMA modeling skipped for this asset.")
            return

    except Exception as e:
        print(f"ARIMA failed: {e}")
        traceback.print_exc()
        return

    preds = np.array(preds)
    true_vals = np.array(true_vals)

    # Evaluation
    rmse = np.sqrt(mean_squared_error(true_vals, preds))
    mae = mean_absolute_error(true_vals, preds)
    hit_rate = np.mean(np.sign(preds) == np.sign(true_vals))

    print(f"RMSE: {rmse:.5f}, MAE: {mae:.5f}, Directional Accuracy: {hit_rate:.2%}")

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(dates, true_vals, label='Actual Log Return')
    plt.plot(dates, preds, label='Forecasted Log Return')
    plt.title(f"{name} - Expanding Window Auto ARIMA Forecast")
    plt.legend()
    metrics_text = f"RMSE: {rmse:.5f}, MAE: {mae:.5f}, Directional Accuracy: {hit_rate:.2%}"
    plt.figtext(0.5, -0.08, metrics_text, ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.7))
    plt.tight_layout()
    plt.show()

# Run for all assets
for name, path in assets.items():
    forecast_asset(name, path)
```

### 4.2.1 Find the order of differencing (d)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import traceback
import os

# File paths (use the signal files generated in Part 3)
base_path = "C:/Users/Dell/Documents/(1) Neu/Big data/Code.py"
assets = {
    "BTC": os.path.join(base_path, "BTC_signal.csv"),
    "VNI": os.path.join(base_path, "VNI_signal.csv"),
    "XAU": os.path.join(base_path, "XAU_signal.csv")
}

# Parameters
forecast_horizon = 1        # Predict next day log return
initial_train_size = 60     # Initial training window (e.g., 60 days)

def adf_test(series, title=''):
    print(f'ADF Test: {title}')
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '# Lags Used', 'Number of Observations Used']
    for value, label in zip(result, labels):
        print(f'{label} : {value}')
    if result[1] <= 0.05:
        print("=> Stationary (reject H0)")
    else:
        print("=> Non-stationary (fail to reject H0)")
    print('-'*40)

def acf_pacf_draw(series, lag_num=40, acf=True, pacf=True, title="", ylim=1):
    num_plots = 1 + int(acf) + int(pacf)
    fig, ax = plt.subplots(1, num_plots, figsize=(12, 6))
    # Plot original or differenced series
    ax[0].plot(series.values.squeeze())
    ax[0].set_title('Series')
    if acf:
        plot_acf(series.values.squeeze(), lags=lag_num, ax=ax[1])
        ax[1].set(ylim=(-ylim, ylim))
        if pacf:
            plot_pacf(series.values.squeeze(), lags=lag_num, ax=ax[2])
            ax[2].set(ylim=(-ylim, ylim))
    elif pacf:
        plot_pacf(series.values.squeeze(), lags=lag_num, ax=ax[1])
        ax[1].set(ylim=(-ylim, ylim))
    fig.suptitle(title)
    plt.show()
```

```python
for asset, path in assets.items():
    print(f"\n=== {asset} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)
    train_data = df['Log_Return'].iloc[:initial_train_size]
    lag_num = min(20, int(len(train_data) // 2) - 1)  # Safe lag number for your sample size

    print("Original Series")
    acf_pacf_draw(train_data, lag_num, True, True, f'{asset} Original Series')
    print("1st Order Differencing")
    acf_pacf_draw(train_data.diff().dropna(), lag_num, True, True, f'{asset} 1st Order Differencing')
    print("2nd Order Differencing")
    acf_pacf_draw(train_data.diff().diff().dropna(), lag_num, True, True, f'{asset} 2nd Order Differencing')
```

**Differencing Order (d) Across Assets**

After examining the ACF and PACF plots for the log return series of BTC, VNI, and XAU, the following common trend is observed:

- *All three assets (BTC, VNI, XAU) have log return series that are already stationary.*
- The ACF and PACF plots for the original series of each asset show that autocorrelations drop quickly and most lags are within the confidence bounds.
- Applying first or second order differencing does not improve stationarity and, in fact, leads to over-differencing, as indicated by erratic spikes and increased noise in the PACF plots.

**Conclusion**

- *The appropriate order of differencing for all three assets is d = 0.*
- This means that the log return series for BTC, VNI, and XAU do not require differencing before ARIMA modeling.
- This trend is consistent across all assets analyzed, indicating that financial log returns are typically stationary and do not benefit from additional differencing.

### 4.2.2 Find the Order of the AR Term (p)

The next step in ARIMA modeling is to determine the number of autoregressive (AR) terms, denoted as \( p \). This is done by inspecting the Partial Autocorrelation Function (PACF) plot of the (stationary) series.

- **Partial autocorrelation** measures the correlation between the series and its lag, after removing the effects of intermediate lags.
- The order \( p \) is typically chosen as the number of lags where the PACF plot shows significant spikes (i.e., values outside the confidence interval).
- For a stationary series, any autocorrelation can be rectified by adding enough AR terms.

**Procedure:**
1. Use the PACF plot of the stationary series (log returns, as determined previously).
2. Count the number of significant lags (spikes outside the confidence bounds).
3. Set \( p \) to this number.

Below is the code to plot the PACF for the stationary series of each asset.

```python
for asset, path in assets.items():
    print(f"\n=== {asset} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)
    train_data = df['Log_Return'].iloc[:initial_train_size]
    lag_num = min(20, int(len(train_data) // 2) - 1)  # Safe lag number for your sample size

    print("PACF of Stationary Series (Log Returns)")
    acf_pacf_draw(train_data, lag_num, acf=False, pacf=True, title=f'{asset} PACF - Log Returns')
```

*Summary Table: Recommended AR (p) Order for Each Asset*

| Asset | Recommended p | Reasoning                                                                 |
|-------|---------------|---------------------------------------------------------------------------|
| BTC   | 1             | Only lag 1 is significant in PACF; all other lags within confidence bounds |
| VNI   | 1             | Only lag 1 is significant in PACF; all other lags within confidence bounds |
| XAU   | 1             | Only lag 1 is significant in PACF; all other lags within confidence bounds |

For all three assets (BTC, VNI, XAU), the PACF plots indicate that an AR(1) model is appropriate. Therefore, set the ARIMA order parameter \( p = 1 \) for each asset.

```python
manual_orders = {
    "BTC": (1, 0, q),
    "VNI": (1, 0, q),
    "XAU": (1, 0, q)  
}
```

### 4.2.4 Arima model diagnostics

```python
import statsmodels.api as sm

def arima_fit(series, order=(1,1,1)):
    # ARIMA model fitting for a given series
    model = sm.tsa.arima.ARIMA(series.values.squeeze(), order=order)
    model = model.fit()
    return model

def get_residual_errors(model):
    # Calculation and drawing the plot residual errors for ARIMA model
    residuals = pd.DataFrame(model.resid)
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    residuals.plot(title="Residuals", ax=ax[0])
    residuals.plot(kind='kde', title='Density', ax=ax[1])
    plt.show()

for asset, path in assets.items():
    print(f"\n=== {asset} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)
    train_data = df['Log_Return'].iloc[:initial_train_size]
    order = manual_orders[asset]

    print(f"Fitting ARIMA model for {asset} with order {order}")
    model = arima_fit(train_data, order=order)
    print(model.summary())

    # Model diagnostics for each asset
    fig = model.plot_diagnostics(figsize=(12, 10))
    plt.show()

    # Plot residual errors for each asset
    get_residual_errors(model)
```

|-------|--------------------------|-------------------------|-------------------------|---------------------------|---------------------------|-------------------------------|
| BTC   | Random, no pattern       | Near normal, fat tails  | Mostly on line, some tail deviation | No significant autocorrelation | Centered at zero, sharp peak | Model assumptions satisfied   |
| VNI   | Random, no pattern       | Near normal, thin tails | Mostly on line, minor tail deviation | No significant autocorrelation | Centered at zero, sharp peak | Model assumptions satisfied   |
| XAU   | Random, no pattern       | Near normal, thin tails | Mostly on line, minor tail deviation | No significant autocorrelation | Centered at zero, sharp peak | Model assumptions satisfied   |

**Conclusion:**  
For all three assets (BTC, VNI, XAU), the ARIMA(1, 0, 0) model residuals:
- Show no significant autocorrelation (white noise).
- Are approximately normally distributed, with some deviations in the tails (expected for financial returns).
- Are centered around zero, with no visible patterns.

The ARIMA(1, 0, 0) model is adequate for modeling the log returns of BTC, VNI, and XAU based on these diagnostics.

### 4.2.5 Arima forecasting

```python
# Define the number of periods in the next quarter (adjust as needed)
next_quarter_steps = 60  # e.g., 60 days for trading data

for asset, path in assets.items():
    print(f"\n=== Next Quarter Forecast for {asset} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)
    full_data = df['Log_Return']

    order = manual_orders[asset]
    model = arima_fit(full_data, order=order)

    # Forecast for the next quarter
    ypred = model.forecast(steps=next_quarter_steps)
    # Calculate simple returns from log returns
    simple_returns = np.exp(ypred) - 1

    # Save forecast to CSV
    forecast_df = pd.DataFrame({
        'Forecast_Step': range(1, next_quarter_steps + 1),
        'Forecasted_Log_Return': ypred,
        'Forecasted_Simple_Return': simple_returns
    })
    forecast_df.to_csv(f"arima_next_quarter_forecast_{asset}.csv", index=False)
    print(f"Saved next quarter forecast for {asset} to arima_next_quarter_forecast_{asset}.csv")
```
