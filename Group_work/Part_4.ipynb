## 4.1 Auto Arima

```python
# PART 4: Time Series Forecasting using ARIMA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import traceback
import os

# File paths (use the signal files generated in Part 3)
base_path = "C:/Users/Dell/Documents/(1) Neu/Big data/Code.py"
assets = {
    "BTC": os.path.join(base_path, "BTC_signal.csv"),
    "VNI": os.path.join(base_path, "VNI_signal.csv"),
    "XAU": os.path.join(base_path, "XAU_signal.csv")
}

# Parameters
forecast_horizon = 60        # Predict next day log return
initial_train_size = 1000     # Initial training window


def forecast_asset(name, path, is_ARIMA=True):
    print(f"\n=== Processing {name} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)

    # Compute log returns
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)

    preds = []
    true_vals = []
    dates = []

    try:
        if is_ARIMA:
            # Automatic tuning of the ARIMA model
            train_data = df['Log_Return'].iloc[:initial_train_size]
            model_auto = auto_arima(
                train_data.values,
                start_p=4,
                start_q=4,
                test='adf',
                max_p=5, max_q=5,
                m=1,
                d=None,
                seasonal=False,
                start_P=0,
                D=0,
                start_Q=0,
                trace=True,
                error_action='ignore',
                suppress_warnings=False,
                stepwise=True
            )
            print(model_auto.summary())
            arima_order = model_auto.order
            print(f"[{name}] Optimal ARIMA order: {arima_order}")

            # Plot diagnostics once
            fig = model_auto.plot_diagnostics(figsize=(12, 10))
            plt.suptitle(f"{name} - Auto ARIMA Diagnostics", fontsize=14)
            plt.show()

            # Expanding window forecast using determined order, forecast every 10th point
            for i in range(initial_train_size, len(df) - forecast_horizon, 10):
                train_data = df['Log_Return'].iloc[:i]
                test_actual = df['Log_Return'].iloc[i + forecast_horizon - 1]
                test_date = df.index[i + forecast_horizon - 1]

                model = ARIMA(train_data, order=arima_order)
                model_fit = model.fit()
                forecast = model_fit.forecast(steps=forecast_horizon)[-1]
                preds.append(forecast)
                true_vals.append(test_actual)
                dates.append(test_date)
        else:
            print("ARIMA modeling skipped for this asset.")
            return

    except Exception as e:
        print(f"ARIMA failed: {e}")
        traceback.print_exc()
        return

    preds = np.array(preds)
    true_vals = np.array(true_vals)

    # Evaluation
    rmse = np.sqrt(mean_squared_error(true_vals, preds))
    mae = mean_absolute_error(true_vals, preds)
    hit_rate = np.mean(np.sign(preds) == np.sign(true_vals))

    print(f"RMSE: {rmse:.5f}, MAE: {mae:.5f}, Directional Accuracy: {hit_rate:.2%}")

# Run for all assets
for name, path in assets.items():
    forecast_asset(name, path)
```

### 4.2.1 Find the order of differencing (d), AR term (p), and MA term (q) for each asset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import traceback
import os

# File paths (use the signal files generated in Part 3)
base_path = "C:/Users/Dell/Documents/(1) Neu/Big data/Code.py"
assets = {
    "BTC": os.path.join(base_path, "BTC_signal.csv"),
    "VNI": os.path.join(base_path, "VNI_signal.csv"),
    "XAU": os.path.join(base_path, "XAU_signal.csv")
}

# Parameters
forecast_horizon = 60        # Predict next day log return
initial_train_size = 1000    # Initial training window

def adf_test(series, title=''):
    print(f'ADF Test: {title}')
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '# Lags Used', 'Number of Observations Used']
    for value, label in zip(result, labels):
        print(f'{label} : {value}')
    if result[1] <= 0.05:
        print("=> Stationary (reject H0)")
    else:
        print("=> Non-stationary (fail to reject H0)")
    print('-'*40)

def acf_pacf_draw(series, lag_num=40, acf=True, pacf=True, title="", ylim=1):
    num_plots = 1 + int(acf) + int(pacf)
    fig, ax = plt.subplots(1, num_plots, figsize=(10, 5))
    # Plot original or differenced series
    ax[0].plot(series.values.squeeze())
    ax[0].set_title('Series')
    if acf:
        plot_acf(series.values.squeeze(), lags=lag_num, ax=ax[1])
        ax[1].set(ylim=(-ylim, ylim))
        if pacf:
            plot_pacf(series.values.squeeze(), lags=lag_num, ax=ax[2])
            ax[2].set(ylim=(-ylim, ylim))
    elif pacf:
        plot_pacf(series.values.squeeze(), lags=lag_num, ax=ax[1])
        ax[1].set(ylim=(-ylim, ylim))
    fig.suptitle(title)
    plt.show()

for asset, path in assets.items():
    print(f"\n=== {asset} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)
    train_data = df['Log_Return'].iloc[:initial_train_size]
    lag_num = min(20, int(len(train_data) // 2) - 1)  # Safe lag number for your sample size

    # Plot all graphs for the current asset together
    fig, axes = plt.subplots(3, 3, figsize=(15, 10))
    fig.suptitle(f'{asset} Analysis', fontsize=16)

    # Original Series
    axes[0, 0].plot(train_data.values.squeeze())
    axes[0, 0].set_title('Original Series')
    plot_acf(train_data.values.squeeze(), lags=lag_num, ax=axes[0, 1])
    plot_pacf(train_data.values.squeeze(), lags=lag_num, ax=axes[0, 2])

    # 1st Order Differencing
    first_diff = train_data.diff().dropna()
    axes[1, 0].plot(first_diff.values.squeeze())
    axes[1, 0].set_title('1st Order Differencing')
    plot_acf(first_diff.values.squeeze(), lags=lag_num, ax=axes[1, 1])
    plot_pacf(first_diff.values.squeeze(), lags=lag_num, ax=axes[1, 2])

    # 2nd Order Differencing
    second_diff = train_data.diff().diff().dropna()
    axes[2, 0].plot(second_diff.values.squeeze())
    axes[2, 0].set_title('2nd Order Differencing')
    plot_acf(second_diff.values.squeeze(), lags=lag_num, ax=axes[2, 1])
    plot_pacf(second_diff.values.squeeze(), lags=lag_num, ax=axes[2, 2])

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    
```

Based on the autocorrelation and partial autocorrelation plots provided for BTC, VNI, and XAU, here is the analysis and suggested ARIMA model orders (p, d, q) for each asset:

| Asset | Differencing Order (d) | AR Order (p) | MA Order (q) |
|-------|------------------------|--------------|--------------|
| BTC   | 0 or 1                 | 1 or 2       | 0 or 1       |
| VNI   | 0 or 1                 | 1 or 2       | 1            |
| XAU   | 0 or 1                 | 1 or 2       | 1 or 2       |

**Notes:**
- The differencing order (d) is determined by the need to make the series stationary.
- The AR order (p) is suggested based on the partial autocorrelation plot, which shows a gradual decline.
- The MA order (q) is suggested based on the autocorrelation plot, which shows a significant spike at lag 1.

```python
manual_orders = {
    "BTC": (1, 0, 0),
    "VNI": (2, 1, 1),
    "XAU": (2, 0, 2)
}
```

### 4.2.2 Arima model training and diagnostics

```python
import statsmodels.api as sm

initial_train_size=1000
def arima_fit(series, order=(1,1,1)):
    # ARIMA model fitting for a given series
    model = sm.tsa.arima.ARIMA(series.values.squeeze(), order=order)
    model = model.fit()
    return model

for asset, path in assets.items():
    print(f"\n=== {asset} ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)
    order = manual_orders[asset]

    # Walk-forward validation setup
    walk_forward_start = initial_train_size
    walk_forward_end = len(df) - forecast_horizon
    true_vals = []
    preds = []
    dates = []
    
    window_size = 250
    for i in range(walk_forward_start, walk_forward_end, 60):
        train_data = df['Log_Return'].iloc[i-window_size:i]
        test_data = df['Log_Return'].iloc[i]
        model = arima_fit(train_data, order=order)
        pred = model.forecast(steps=1)[0]
        preds.append(pred)
        true_vals.append(test_data)
        dates.append(df.index[i])

    # Convert to numpy arrays for metrics
    true_vals = np.array(true_vals)
    preds = np.array(preds)
    dates = np.array(dates)

    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(true_vals, preds))
    mae = mean_absolute_error(true_vals, preds)
    hit_rate = np.mean(np.sign(preds) == np.sign(true_vals))

    # Model diagnostics for the last fitted model
    print(f"RMSE: {rmse:.5f}, MAE: {mae:.5f}, Directional Accuracy: {hit_rate:.2%}")
    print("Displaying ARIMA model diagnostics for the last fitted model window:")
    model.plot_diagnostics(figsize=(10, 8))
    plt.tight_layout()
    plt.show()
    plt.close()
    
    # Plotting
    plt.figure(figsize=(8, 4))
    plt.plot(dates, true_vals, label='Actual Log Return', color='blue')
    plt.plot(dates, preds, label='Forecasted Log Return', color='orange', linestyle='--')
    plt.title(f"{asset} - Walk-Forward ARIMAX Forecast (Order: {manual_orders[asset]})")
    plt.xlabel('Date')
    plt.ylabel('Log Return')
    plt.legend()
    metrics_text = f"RMSE: {rmse:.5f}, MAE: {mae:.5f}, Directional Accuracy: {hit_rate:.2%}"
    plt.figtext(0.5, -0.08, metrics_text, ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.7))
    plt.tight_layout()
    plt.show()
    plt.close()
```

***Model Diagnostics (Residual Analysis)***

*Standardized Residuals*
- The standardized residual plots for all assets (BTC, VNI, XAU) show that residuals fluctuate around zero, which is a good sign. However, there are some spikes, indicating occasional outliers or periods where the model does not fit well.

*Histogram plus Estimated Density*
- The residual histograms are roughly bell-shaped and centered at zero, with the KDE (Kernel Density Estimate) curve closely following the normal distribution curve (N(0,1)). This suggests that the residuals are approximately normally distributed, which is a key assumption for ARIMA models.
- There are some deviations in the tails, indicating occasional extreme values (fat tails), which is common in financial time series.

*Normal Q-Q Plot*
- The Q-Q plots show that most points lie on the straight line, indicating that the residuals are close to normal. However, there are deviations at the ends (tails), again suggesting some outliers or heavy tails.

*Correlogram (ACF of Residuals)*
- The autocorrelation plots (correlograms) show that most autocorrelations are within the confidence bounds, indicating that the residuals are largely uncorrelated (white noise). This means the ARIMA model has captured most of the autocorrelation structure in the data.
- The first lag is always 1 (by definition), but subsequent lags are close to zero.

***Walk-forward forcast performance***

| Asset | Model behavior |
|-------|-----------------------------------|
| **BTC** | The forecasted log returns are much smoother than the actual returns and do not capture the volatility spikes well. This is typical for ARIMA models on financial returns, which are hard to predict due to their noise and volatility clustering. |
| **VNI** | Similar to BTC, the model produces smooth forecasts and misses large jumps in the actual returns. |
| **XAU** | The model again produces smooth forecasts, with slightly better directional accuracy. |

- Model Fit: The ARIMA models are generally well-specified: residuals are close to white noise and approximately normal, indicating that the models have captured the main autocorrelation structure.
- Forecasting Power: The models struggle to capture the volatility and large jumps in actual returns, which is expected for ARIMA models on financial return series.
- Directional Accuracy: Directional accuracy is above 50% for all assets, which is better than random guessing, but not by a large margin.

### 4.2.3 Arima forecasting

```python
# --- Forecasting the Next Quarter ---

forecast_horizon = 60  # Number of days in the next quarter (adjust if needed)

for asset, path in assets.items():
    print(f"\n=== {asset} - Next Quarter Forecast ===")
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True)
    df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
    df.dropna(inplace=True)
    order = manual_orders[asset]

    # Use the last window_size points for training
    window_size = 250
    last_train_data = df['Log_Return'].iloc[-window_size:]
    model = arima_fit(last_train_data, order=order)
    forecast = model.forecast(steps=forecast_horizon)

    # Create a date range for the forecast
    last_date = df.index[-1]
    forecast_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=forecast_horizon, freq='B')

    # Plot the forecast
    plt.figure(figsize=(10, 5))
    plt.plot(df.index[-window_size:], last_train_data, label='Recent Log Return', color='blue')
    plt.plot(forecast_dates, forecast, label='Forecasted Log Return (Next Quarter)', color='red', linestyle='--')
    plt.title(f"{asset} - Next Quarter Log Return Forecast (Order: {order})")
    plt.xlabel('Date')
    plt.ylabel('Log Return')
    plt.legend()
    plt.tight_layout()
    plt.show()
    plt.close()
```

#### Forecast Results Analysis
**Forecast Plots**
- For all assets (BTC, VNI, XAU), the ARIMA models forecast the next quarter's log returns as nearly flat lines, close to zero.
- This is a typical outcome for ARIMA models on financial return series, especially when the series is close to white noise (as your residual diagnostics suggest).
- The recent actual log returns are volatile, but the forecasted values show little to no variation, indicating the model expects mean-reverting or unpredictable behavior.

**Forecasting & Realistic Assumptions**
- The ARIMA models assume linear relationships and stationarity, which are reasonable first steps for financial time series.
- The flat forecasts reflect the statistical reality that, after accounting for autocorrelation, there is little predictable structure left in daily returns for these assets.

**Conclusion**
The results highlight the inherent unpredictability and noise in financial return series, justifying the need for more sophisticated models or alternative approaches in subsequent parts of this project.
