### 7.5.Summary of conclusions & proposed investment strategies

```python
# Calculate the Kelly Ratio for BTC, VNI, XAU
import pandas as pd

kelly_fraction = pd.DataFrame({
    "Asset": ["BTC", "VNI", "XAU"],
    "Kelly_Ratio": [
        btc_data["Return_1"].mean()/btc_data["Return_1"].std(),
        vni_data["Return_1"].mean()/vni_data["Return_1"].std(),
        xau_data["Return_1"].mean()/xau_data["Return_1"].std()
    ]
})

print("Kelly Ratio cho từng tài sản:")
print(kelly_fraction)
```

- BTC has a high Kelly Ratio but is highly volatile → Suitable for risk-seeking investors, can generate significant profits but comes with high risks. 
- VNI has an average Kelly Ratio → Stable but does not surge strongly, suitable for long-term investment, avoiding high risks. 
- XAU has a low Kelly Ratio but is less influenced by news → It is a good defensive asset, helping to protect the portfolio when the market is volatile.

* Practical applications: 
- Risk-loving investors → May focus on BTC with a higher allocation. 
- Safety-seeking investors → XAU is a good choice to reduce portfolio volatility. 
- Long-term investors → VNI aligns with a strategy of stable investment and sustainable growth.
